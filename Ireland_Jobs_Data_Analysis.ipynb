```python
import requests
import pandas as pd
import time
```

```python
# 🔐 PASTE YOUR RAPIDAPI KEY
RAPIDAPI_KEY = "5369d44daemshe43ddc12bc9c0ddp166696jsnc6049a7721d7"

# JSearch config (DO NOT CHANGE)
RAPIDAPI_HOST = "jsearch.p.rapidapi.com"
BASE_URL = "https://jsearch.p.rapidapi.com/search"

HEADERS = {
    "x-rapidapi-key": RAPIDAPI_KEY,
    "x-rapidapi-host": RAPIDAPI_HOST
}

# Limits (free tier safe)
MAX_PAGES = 3
SLEEP_SECONDS = 1
```

```python
ROLE_QUERIES = {
    "Software Engineer": "software",
    "Data Analyst": "data",
    "Product Manager": "product",
    "DevOps Engineer": "devops",
    "QA Engineer": "qa",
    "Business Analyst": "business analyst",
    "Electrical Engineer": "electrical",
    "Mechanical Engineer": "mechanical"
}
```

```python
def fetch_jobs_for_role_ie(query, max_pages=MAX_PAGES):
    all_jobs = []

    for page in range(1, max_pages + 1):
        params = {
            "query": query,
            "location": "Ireland",        # primary
            "page": page,
            "num_pages": 1,
            "remote_jobs_only": False
        }

        response = requests.get(BASE_URL, headers=HEADERS, params=params)
        print(f"   Page {page} | Status {response.status_code}")

        if response.status_code != 200:
            print("   ❌ Error:", response.text[:200])
            break

        data = response.json().get("data", [])
        print(f"      → {len(data)} jobs")

        if not data:
            break

        all_jobs.extend(data)
        time.sleep(SLEEP_SECONDS)

    # 🔁 Fallback: Dublin only (Ireland jobs are often Dublin-tagged)
    if len(all_jobs) == 0:
        print("   ⚠️ No jobs found. Trying Dublin fallback...")
        for page in range(1, max_pages + 1):
            params = {
                "query": query,
                "location": "Dublin, Ireland",
                "page": page,
                "num_pages": 1
            }

            response = requests.get(BASE_URL, headers=HEADERS, params=params)
            data = response.json().get("data", [])
            print(f"      Dublin → {len(data)} jobs")

            if not data:
                break

            all_jobs.extend(data)
            time.sleep(SLEEP_SECONDS)

    return all_jobs
```

```python
results = []

for role, query in ROLE_QUERIES.items():
    print(f"\n🔄 Fetching jobs for: {role}")
    jobs = fetch_jobs_for_role_ie(query)

    results.append({
        "job_title": role,
        "open_roles": len(jobs)
    })

df_ie_roles = (
    pd.DataFrame(results)
    .sort_values("open_roles", ascending=False)
    .reset_index(drop=True)
)

df_ie_roles
```

```python
df_ie_roles.to_csv("ireland_open_roles_by_title.csv", index=False)
"ireland_open_roles_by_title.csv"
```
